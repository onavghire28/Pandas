```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("C:/Users/Om/Downloads/imdb-top-1000 (1).csv")
movies.head(5)
```

```python
#Group by

genres = movies.groupby('Genre')['Gross'].sum()
genres
```

```python
#Find the genre which having high avg Rating

avg = movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending = False).head(1)
avg
```

```python
#Find the popolar director top 5

popular = movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending = False).head(5)
popular
```
