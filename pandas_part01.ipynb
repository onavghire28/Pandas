```python
#Series with list

import pandas as pd
import numpy as np

state = ['Maharashtra','Goa','Gujrat','Rajastan']
pd.Series(state)
```

```python
#Custom index we can provided

marks = [90,45,72,94,50]
subject = ['Marathi','Hindi','English','Science','Math']

pd.Series(marks, index= subject)
```

```python
#Series with dictonory

Mark = {
    "Marathi":90,
    "Hindi":75,
    "Science":71,
    "English":73,
    "Math":80
}

pd.Series(Mark)
```
